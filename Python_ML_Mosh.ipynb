```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

music = pd.read_csv ('/content/drive/MyDrive/music.csv.zip (Unzipped Files)/music.csv')
X = music.drop(columns = ['genre'])
Y = music ['genre']
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2)

#print (X)
#print (Y)

model = DecisionTreeClassifier()                   #creating an instance 
model.fit(X_train,Y_train)                                      #setting the two datasets as parametres
predictions = model.predict(X_test)     #making the model predict
#predictions 

score = accuracy_score(Y_test, predictions)    #determining accuracy of model
score
```
